```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Question 1, part 1

N = 200 
x = np.array([float(i-1)/(N-1) for i in range(1, N+1)])
f_x = np.sin(2*np.pi*x)
gwn = np.random.normal(0, 0.1, size=N)
y = f_x + gwn

plt.plot(x, f_x)
plt.plot(x, y, 'bo')
```

```python
# Question 1, part 2

coefficients = []
for regularization in [0, 3, 6, 9, 12]:
    coefficients.append([])
    for i,d in enumerate([0, 1, 3, 5, 9]):
        coefficients[-1].append([])
        identity = np.identity(d+1)
        MSE = []
        for size in [10, 50, 100, 150, 200]:
            indices = np.sort(np.random.choice(200, size, replace=False))
            xtrain = x[indices]
            ytrain = y[indices]
            
            X = np.ones((size, 1))
            for k in range(1, d+1):
                X = np.c_[X,xtrain**k]
            
            W = X.T.dot(X)
            W += regularization*identity
            W = np.linalg.inv(W)
            
            XtY = X.T.dot(ytrain)
            W = W.dot(XtY)
            coefficients[-1][-1].append(W)
            prediction = X.dot(W)
            MSE.append(np.mean((prediction-ytrain)**2))
            
        plt.title('Lambda = ' + str(regularization))
        plt.plot([10, 50, 100, 150, 200], MSE)
        plt.legend(["order = 0", "order = 1", "order = 3", "order = 5", "order = 9"])

    plt.show()
```
